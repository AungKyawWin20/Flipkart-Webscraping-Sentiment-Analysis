```python
import pandas as pd

df = pd.read_csv('flipkart_scraped_data.csv')
df.info()
```

```python
df['Reviews'] = df['Reviews'].astype(str).str.split('|')
df_exploded = df.explode('Reviews').reset_index(drop=True)

df_exploded['Reviews'] = df_exploded['Reviews'].str.strip()
df_exploded = df_exploded[df_exploded['Reviews'] != ""]
```

```python
df_exploded.info()
```

```python
df_exploded.isnull().sum()
```

```python
df_exploded.dropna(inplace=True)
```

```python
df_exploded.head()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
model_name = "siebert/sentiment-roberta-large-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
import torch
```

```python
model.eval()
device="cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
sentiment=[]
with torch.no_grad():
    for text in df_exploded['Reviews']:
      inputs=tokenizer(text,return_tensors='pt',padding=True,truncation=True,max_length=100)
      inputs={k:v.to(device) for k,v in inputs.items()}
      outputs=model(**inputs)
      values=torch.nn.functional.softmax(outputs.logits,dim=-1)
      final_value=torch.argmax(values)
      result="positive" if final_value==1 else "negative"
      sentiment.append(result)
print(sentiment)
```

```python
df_exploded['Sentiment'] = sentiment
df_exploded.to_csv('flipkart_data_with_sentiment.csv', index=False)
```
